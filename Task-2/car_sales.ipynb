Thus we can conclude that there is no null values from the given dataset. Then here will be analysing the data about it's structure and segregating them with dependent variable and independent-variables.

```python
df.head(5)
```

Here we can find the categorical values like : Car_name , fuel_type , seller_type , transmission..etc
Now let's convert categorical values into numerical values

```python
x = ['Car_Name', 'Fuel_Type', 'Seller_Type', 'Transmission']
df = pd.get_dummies(df , columns = x , drop_first=True )
df.head(5)
```

```python
# now to find the age of the car let's do some mathematical analysis
# current age = new_car_bought_year - 2024(current_year)
df['Car_Age'] = 2024 - df['Year']
df.drop('Year' , axis=1 , inplace=True)
df.describe()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score

x = df.drop('Selling_Price', axis=1) 
y = df['Selling_Price']   


# model making
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=0.5 , random_state= 34)

model = RandomForestRegressor(n_estimators=100 , random_state=34)
model.fit(x_train , y_train)

predict = model.predict(x_test)
print(f'RMSE: {mean_squared_error(y_test, predict, squared=False)}')
print(f'R² score: {r2_score(y_test, predict)}')


```

```python
import streamlit as st
import pandas as pd
import pickle


model = pickle.load(open('model.sav', 'rb'))

def run():
    st.title('Selling Price Prediction')
    feature1 = st.number_input('Feature 1')
    feature2 = st.number_input('Feature 2')
    if st.button('Predict'):
        input_data = pd.DataFrame([[feature1, feature2]], columns=['feature1', 'feature2'])
        prediction = model.predict(input_data)
        st.success(f'The predicted selling price is: {prediction[0]}')

# if __name__ == '__main__':
run()
```
