```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HousingData.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
mea = df['CRIM'].mean()
mea
df['CRIM'].fillna(value = mea , inplace=True)
df.isnull().sum()
```

```python
zn_mea = df['ZN'].mean()
df['ZN'].fillna(value = mea , inplace=True)
indus_mean = df['INDUS'].mean()
df['INDUS'].fillna(value=indus_mean , inplace=True)
chas_mean = df['CHAS'].mean()
df['CHAS'].fillna(value=chas_mean , inplace=True)
age_mean = df['AGE'].mean()
df['AGE'].fillna(value=age_mean , inplace=True)
ls_mean = df['LSTAT'].mean()
df['LSTAT'].fillna(value= ls_mean , inplace=True)
print('The number null values after preprocessing :')
df.isnull().sum()
```

```python
print("checking the feature name : ")
print(df.info)
```

```python
df.describe()
```

MODEL DEVLOPMENT

```python
x = df[['RM' , 'PTRATIO' , 'AGE']]
y = df['LSTAT']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test  = train_test_split( x , y , test_size= 0.2 , random_state=42)
```

```python
from sklearn import linear_model
ln = linear_model.LinearRegression()
ln.fit(x_train , y_train)
```

```python
predict = ln.predict(x_test)
```

```python
plt.scatter(y_test , predict)
plt.xlabel('y Test')
plt.ylabel('Predictions')
```

Evaluating the Model

```python
from sklearn import metrics
metrics_of_meansqaure = metrics.mean_squared_error(y_test , predict)
rmse = np.sqrt(metrics.mean_squared_error(y_test , predict))
print("meansquare : " , metrics_of_meansqaure)
print("rmse : " , rmse)
```

```python
# metrics_of_mean
sns.distplot((y_test , predict),bins=20)
```
